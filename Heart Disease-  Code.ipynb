What is the average resting blood presure?

There are a wide range of values over ages.

So will create a function to get the average value per age.

```python
# Function that will take missing values, get the average for THAT age and then fill in the missing value.
def fill_missing_resting_blood_pressure(data_df):
    
    # Group dataframe by age and calculate the average resting_blood_pressure
    age_mean_rbp = data_df.groupby("age")["resting_blood_pressure"].mean()
    
    # check if there is null value in resting blood_pressure
    # if there is replace it for that age with the resting_blood_pressure in age_mean_rbp
    for i, row in data_df.iterrows():
        if pd.isna(row["resting_blood_pressure"]):
            age = row["age"]
            if age in age_mean_rbp:
                mean_rbp = age_mean_rbp[age]
                data_df.at[i, "resting_blood_pressure"] = mean_rbp

# call function
fill_missing_resting_blood_pressure(data_df)
```

```python
# Check if missing values filled
num_missing_per_age_update = data_df.groupby("age")["resting_blood_pressure"].apply(lambda x: x.isna().sum())
# num_missing_per_age_update
```

No longer missing values and we have filled in the missing values with approtiate values by age rather than an overall average.

### Univariate Selection

Statistical tests can be used also to detect which features have the strongest relationships with the output variables.

The Scikit-learn library provides the SelectKBest class that has a number of statistical tests that can be used to select the number of variables.

One option is to use the scoring function f_classif which measures the degree of linear dependence between the features and target variables. Higher vales indicate a stronger linear relationship.

```python
# Use SelectKBest to find the 7 best features.
best_features = SelectKBest(score_func=f_classif, k=7)
fit = best_features.fit(data_df, y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(data_df.columns)

#print(dfscores); print(dfcolumns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  
print(featureScores.nlargest(10,'Score'))  
```

## Create a user input and make a predication

I want to create a function, there a user will be asked questions and then a model will be run to give a prediction if a person has heart disease.

```python
# function to get answers to questions

def user_input():
    # create a list to store the answers
    user_answers = []
    
    # 13 questions to ask to represet the 13 columns.
    # Later can add in some error control. But for this project as a concept we have not included it.
    questions = [
        "1. Age in years(input number, eg. 42): " ,
        "2. Sex (input 1 for male, 0 for female, eg. 1): ",
        "3. Any Chest Pain (input 1 for typical angina, 2 for atypical angina, 3 for non-anginal pain, 4 for asymptomatic, eg. 3): ",
        "4. Resting blood presure (input in mm Hg, eg.142): ",
        "5. Serum Colesterol (input in mg/dl, eg.157): ",
        "6. Fasting blood suger > 120 mg/dl (input 1 for ture, if greater than 120, 0 for false, if less than 120, eg. 0): ",
        "7. Resting ECG results (input 0 if normal, 1 have ST-T wave, 2 if left ventricular hypertropy, eg. 0): ",
        "8. Maximum heart rate (input number, eg. 150): ",
        "9. Exercise induced angina (input 1 for Yes, 0 for No, eg. 0): ",
        "10. ST depression induced by exercise relative to rest ( eg. 0 ): ",
        "11. Peak Exerise ST segment (input 1 for upsloping, 2 for flat, 3 for downsloping, eg. 2): ",
        "12. Thalassemia ( input 3 for normal, 6 for fixed defect, 7 for reverisble defect, eg. 3): "
    ]
    
    # loop though the questions and store answers
    for question in questions:
        answer = input(question)
        
        # need to store the object types to the correct type.
        if question.startswith("Age") or question.startswith("Resting Blood") or question.startswith("Serum") or question.startswith("Maximum") or question.startswith("ST") or question.startswith("Age"):
            answer = float(answer)
        else:
            answer = int(answer)
        
        user_answers.append(answer)
    return user_answers


# get answers from the users
user_answers = user_input()

print(user_answers)
            
```
